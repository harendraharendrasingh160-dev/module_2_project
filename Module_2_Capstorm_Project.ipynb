## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Drop columns which are not important
data.drop(["ID","SSN","Type_of_Loan","Credit_Mix"],axis=1,inplace=True)
````

````python
# Make new column of mean and mode value of various parameter with refrence of customer id
data["Avg_delay_from_due_date"] = data.groupby("Customer_ID")["Delay_from_due_date"].transform("mean")
data["Avg_no._of_due_date"] = data.groupby("Customer_ID")["Num_of_Delayed_Payment"].transform("mean")
data["Avg_histry_age"] = data.groupby("Customer_ID")["Credit_History_Age"].transform("mean")
data["Avg_total_EMI"] = data.groupby("Customer_ID")["Total_EMI_per_month"].transform("mean")
data["Avg_monthly_balance"] = data.groupby("Customer_ID")["Monthly_Balance"].transform("mean")
data["Avg_credit_utilization_ratio"] = data.groupby("Customer_ID")["Credit_Utilization_Ratio"].transform("mean")
data["Customer_Credit_Mode"] = data.groupby("Customer_ID")["Credit_Score"] \
                                   .transform(lambda x: x.mode()[0])
````

````python
# calculate ratio which are helpful for analysis and put in new column
data['EMI_to_Salary_Ratio'] = data['Total_EMI_per_month'] / data['Monthly_Inhand_Salary']
data['Debt_to_Income_Ratio'] = data['Outstanding_Debt'] / data['Annual_Income']
data['Savings_Ratio'] = data['Monthly_Balance'] / data['Monthly_Inhand_Salary']
data['Loan_per_Account'] = data['Num_of_Loan'] / data['Num_Bank_Accounts']
delay_sum = data.groupby("Customer_ID")["Num_of_Delayed_Payment"].transform("sum")
month_count = data.groupby("Customer_ID")["Month"].transform("count")
data["Delay_Ratio"] = delay_sum / month_count
data['Credit_Age_Ratio'] = data['Avg_histry_age'] / (data['Age'] * 12)
data['Credit_History_Years'] = data['Avg_histry_age'] / 12
````

````python
# Get a information about ratios for perticular customer
Customer_ID_Number = int(input("Enter Customer ID Number: "))
print(f'Name of customer : {data[data["Customer_ID"]==Customer_ID_Number]["Name"].iloc[0]}')
print(f'Credit Utilization Ratio : {round(data[data["Customer_ID"]==Customer_ID_Number]["Credit_Utilization_Ratio"].mean(),2)}')
print(f'EMI to Salary Ratio : {round(data[data["Customer_ID"]==Customer_ID_Number]["EMI_to_Salary_Ratio"].mean(),2)}')
print(f'Debt to Income Ratio : {round(data[data["Customer_ID"]==Customer_ID_Number]["Debt_to_Income_Ratio"].mean(),2)}')
print(f'Savings Ratio : {round(data[data["Customer_ID"]==Customer_ID_Number]["Savings_Ratio"].mean(),2)}')
print(f'Loan per Account : {round(data[data["Customer_ID"]==Customer_ID_Number]["Loan_per_Account"].mean(),2)}')
print(f'Delay Ratio : {round(data[data["Customer_ID"]==Customer_ID_Number]["Delay_Ratio"].mean(),2)}')
print(f'Credit Age Ratio : {round(data[data["Customer_ID"]==Customer_ID_Number]["Credit_Age_Ratio"].mean(),2)}')
print(f'Credit History Years : {round(data[data["Customer_ID"]==Customer_ID_Number]["Credit_History_Years"].mean(),2)}')
print(f'Credit Score : {data[data["Customer_ID"]==Customer_ID_Number]["Credit_Score"].mode().iloc[0]}')
````

````python
# Get whole rows for perticular customer with respect to months
Customer_ID_Number = int(input("Enter Customer ID Number: "))
Start_month = int(input("Enter Start month: "))
End_month = int(input("Enter End month: "))

def get_customer_data(customer_id, start_month, end_month):

    filtered = data[
        (data["Customer_ID"] == customer_id) &
        (data["Month"] >= start_month) &
        (data["Month"] <= end_month)
    ]

    if not filtered.empty:
        return filtered
    else:
        return "No data found"

get_customer_data(Customer_ID_Number, Start_month, End_month)
````

This analysis explored customer financial behavior to understand the key drivers of credit risk and credit score classification. The dataset revealed that payment discipline, credit utilization, and financial ratios play a crucial role in determining a customer’s creditworthiness.

Customers with higher delays in payment ( Delay from due date  and  Num of Delayed Payment ) consistently showed lower credit scores, making payment history the most influential factor. Correlation and pair plot analysis further confirmed that delay-related variables and debt-related ratios are strong predictors of credit risk, while savings and controlled utilization act as stabilizing factors.

Overall, the study concludes that a combination of behavioral (payment patterns) and financial (income, debt, savings) indicators can effectively predict credit score.
